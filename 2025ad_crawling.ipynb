```python
import pandas as pd
import requests

import random
from tqdm import tqdm
import datetime
import time
from selenium.webdriver.support.ui import WebDriverWait
from tqdm import tqdm
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
driver = webdriver.Chrome('./chromedriver.exe')

url="https://www.2025ad.com/search-results"

driver.get(url)

searchText_T = ["self-driving", "autonomous", "driverless"]

driver.find_element_by_xpath('/html/body/main/div/div[1]/div/div/input').send_keys(searchText_T[2])
driver.find_element_by_xpath('/html/body/main/div/div[1]/div/div/button').click()

news_urls=[]
news_contents=[]
news_titles=[]

for j in range(1, 10) :
    if j==1 :
        pass
    
    elif j==2 :
        driver.find_element_by_xpath('//*[@id="hs_cos_wrapper_search_results"]/div/div/a').click()
        time.sleep(1)
        
    else :
        driver.find_element_by_xpath('//*[@id="hs_cos_wrapper_module_1520438261786150"]/div/div/a[2]').click()
        
    for i in tqdm(range(1,11)) :
            try :
                news_title = driver.find_element_by_xpath('//*[@id="hsresults"]/li['+ str(i) +']/a')

                time.sleep(1)

                url = news_title.get_attribute('href')

                news_urls.append(url)
    
            except:
                pass
            
            time.sleep(3)
            
            a = driver.find_element_by_xpath('//*[@id="hsresults"]/li['+str(i)+']').click()
            html = driver.page_source
            soup = BeautifulSoup(html,'html.parser')
            
            time.sleep(2)

            title = soup.find('h1',{'class':'article-page__title'})
            
            time.sleep(1)
            
            content = soup.find('div',{'class':'article-page__content-wrap l-span12'})
            
            if content is None:
                news_contents.append('N/A')
            
            else:
                contents = content.text.strip()
                news_contents.append(contents)
            time.sleep(1)

            if title is None:
                news_titles.append('N/A')
            
            else:    
                titles = title.text.strip()
                news_titles.append(titles)



            driver.back()
```

```python
news_urls = pd.DataFrame(news_urls[:])

news_titles = pd.DataFrame(news_titles[:])

news_contents = pd.DataFrame(news_contents[:])



result3 = pd.concat([news_urls, news_titles, news_contents],axis=1)

result3.columns=['urls', 'titles','contents']

result=result3.reset_index(drop= True)

result.to_csv('news_crawl.csv',index=False)
```
